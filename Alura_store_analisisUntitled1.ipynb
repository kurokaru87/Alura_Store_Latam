```python
import matplotlib.pyplot as plt

def graficar_categorias(serie, titulo):
    plt.figure(figsize=(10,5))
    serie_sorted = serie.sort_values(ascending=False)  # Ordenar de mayor a menor
    plt.bar(serie_sorted.index, serie_sorted.values)
    plt.title(titulo)
    plt.xlabel("Categoría")
    plt.ylabel("Cantidad Vendida")
    plt.xticks(rotation=45)
    plt.show()

# Gráficos por tienda
graficar_categorias(cat_t1, "Categorías más vendidas - Tienda 1")
graficar_categorias(cat_t2, "Categorías más vendidas - Tienda 2")
graficar_categorias(cat_t3, "Categorías más vendidas - Tienda 3")
graficar_categorias(cat_t4, "Categorías más vendidas - Tienda 4")

```

```python
#calculo del costo de envio promedio de cada tienda.

envio_t1 = t1["Costo de envÃ­o"].mean()
envio_t2 = t2["Costo de envÃ­o"].mean()
envio_t3 = t3["Costo de envÃ­o"].mean()
envio_t4 = t4["Costo de envÃ­o"].mean()

envio_promedio = {
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Costo de Envío Promedio": [envio_t1, envio_t2, envio_t3, envio_t4]
}

import pandas as pd
df_envio = pd.DataFrame(envio_promedio)
df_envio

df_envio["Costo de Envío Promedio"] = df_envio["Costo de Envío Promedio"].round(2)
df_envio



```

```python
import matplotlib.pyplot as plt
import numpy as np

# Valores del costo de envío promedio
values = df_envio["Costo de Envío Promedio"].tolist()
labels = df_envio["Tienda"].tolist()

# Cálculo acumulativo
cumulative = np.cumsum([0] + values[:-1])

# Colores: verde si sube, rojo si baja
colors = ['green' if val >= 0 else 'red' for val in values]

plt.figure(figsize=(10,6))
plt.bar(labels, values, bottom=cumulative, color=colors, edgecolor='black')

# Agregar valores numéricos en las barras
for i in range(len(values)):
    plt.text(i, cumulative[i] + values[i]/2, f"{values[i]:.2f}",
             ha='center', va='center', color='white', fontsize=10, fontweight='bold')

plt.title("Gráfico de Cascada - Costo de Envío Promedio por Tienda")
plt.ylabel("Costo de Envío Promedio")
plt.show()
```

Gráfico de Cascada: Costo de Envío Promedio

Se realizó un gráfico de cascada para visualizar cómo varía el costo de envío promedio entre las tiendas.  
Este tipo de gráfico permite observar incrementos o reducciones de manera progresiva, facilitando la comparación directa de costos entre cada punto de venta.
